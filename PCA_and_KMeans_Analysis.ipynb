```python
# A)

import numpy as np
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris()
X = iris.data

# Perform PCA for 1, 2, 3, and 4 components
for k in range(1, 5):
    pca = PCA(n_components=k)
    pca.fit(X)
    print(f"Number of components: {k}")
    print(f"Fraction of variation explained: {sum(pca.explained_variance_ratio_):.3f}")
```

```python
#B)

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Perform PCA for 1, 2, 3, and 4 components
for k in range(1, 5):
    pca = PCA(n_components=k)
    pca.fit(X_scaled)
    print(f"Number of components: {k}")
    print(f"Fraction of variation explained: {sum(pca.explained_variance_ratio_):.3f}")
```

```python
#C)

import matplotlib.pyplot as plt

X_raw, y_raw = load_iris(return_X_y=True)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_raw)

pca = PCA(n_components=2)
X_proj = pca.fit_transform(X_scaled)

plt.scatter(X_proj[:,0], X_proj[:,1], c=y_raw, cmap='viridis')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.show()
```

```python
#D)

from sklearn.cluster import KMeans

# Instantiate the KMeans model with 3 clusters
kmeans = KMeans(n_clusters=3)

# Fit the model to the projected data
kmeans.fit(X_proj)

# Get the cluster assignments for each instance
assignments = kmeans.labels_

# Visualize the clusters
plt.scatter(X_proj[:,0], X_proj[:,1], c=assignments, cmap='viridis')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.show()
```

 The KMeans algorithm is able to cluster the data into three groups based on their proximity to the centroids of each cluster.
